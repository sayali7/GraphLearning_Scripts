```python
import os
import time
import random
import pickle
import tqdm

import pandas as pd
import numpy as np
import grape
from grape import edge_list_utils, Graph
from grape.embedders import Node2VecCBOWEnsmallen
```

```python
GRAPH_EMB_FOLDER="./../Graphs/"
```

```python
file = open(GRAPH_EMB_FOLDER+'sample_id.pkl', 'rb')
sample_ids = pickle.load(file)
```

```python
file = open(GRAPH_EMB_FOLDER+'graph_with_edgeW.pkl', 'rb')
graphs = pickle.load(file)
```

```python
graphs
```

```python
all_emb=[]
```

```python
for graph in tqdm.tqdm(graphs):
    #edges_df = sample[['from_gene','to_gene']]
    edges_df = pd.DataFrame(graph.edge_index).T
    edges_df.columns = ["from", "to"]
    
    graph = Graph.from_pd(edges_df=edges_df, edge_src_column="from", 
                          edge_dst_column="to", directed=False)
    
    emb = Node2VecCBOWEnsmallen(embedding_size = 64, epochs = 30, walk_length = 128
                           ).fit_transform(graph)
    
    node_emb = emb.get_all_node_embedding()
    all_emb.append(node_emb)
```

```python
with open(GRAPH_EMB_FOLDER+'homo_c15_5TF_10Tar_p2_5_graph_with_edgeW_subgraphs.pkl', "wb") as f:
    pickle.dump(all_emb, f)
```
